```python
import pandas as pd
house_price = pd.read_csv('/content/Housing.csv')
house_price
```

```python
house_price.replace({'mainroad' : {'yes' : 0, 'no' : 1}, 'guestroom' : {'yes' : 0, 'no' : 1}, 'basement' : {'yes' : 0, 'no' : 1}, 'hotwaterheating' : {'yes' : 0, 'no' : 1}, 'airconditioning' : {'yes' : 0, 'no' : 1}, 'prefarea' : {'yes' : 0, 'no' : 1}, 'furnishingstatus' : {'furnished' : 0, 'semi-furnished' : 1, 'unfurnished' : 2}}, inplace = True)
#house_price.replace({'mainroad' : {'yes' : 0, 'no' : 1}, 'guestroom' : {'yes' : 0, 'no' : 1}}, inplace = True)
```

```python
x = house_price.drop(['price'], axis = 1)
y = house_price['price']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn import metrics
y_predict = lr.predict(x_test)
target = pd.DataFrame({'Actual Value' :y_test, 'Predicted Value' : y_test})
target
```

```python
from sklearn.metrics import r2_score, max_error, mean_absolute_error
accuracy =  r2_score(y_test, y_predict)
print('Accuracy = ', accuracy)
mean_absolute_error = mean_absolute_error(y_test, y_predict)
print('Mean absolute error = ', mean_absolute_error)
max_error = max_error(y_test, y_predict)
print('Max Error = ', max_error)
```
